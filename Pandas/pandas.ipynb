## Pandas
* It provides data structures like DataFrames and Series, which allow for efficient handling and processing of structured data, similar to tables in SQL or spreadsheets.

```python
import pandas as pd
```

## Key data sturctures in pandas
* Series 1-d data structure capable for storing data of any datatypes
* DataFrame 2-d data structure

# Series

```python
#using list--------

sr=pd.Series([1,2,3,4,5])
print('default index-->\n',sr)

sr_indx=pd.Series([1,2,3,4,5],index=["q","p","r","s","t"],dtype=float, name="pyhton") 
#name stands for name of the data
#dtype stands for datatype

print('\ncustome index-->\n',sr_indx)

print('\n',type(sr)) #to know the type of data

#using dictionary------
sr_dic={"name":['nnn','ttt','rrr'],"marks":[12,34,55]}
var=pd.Series(sr_dic)
print("\ndic data--->\n",var)
```

* Use of series beneficial instead of numpy why?

```python
s1=pd.Series(12,index=[1,2,3,4])
s2=pd.Series(12,index=[1,2,3,4,5,6,7,8])

print(s1+s2)

#here it works and give Nan at the place of missing data whereas numpy it generates error!!
```

# Dataframe

```python
dic={"name":["ds","djs","dws","hdhdw","jwdww","ueuwue"],"roll no":[23,45,45,74,94,48]}
df=pd.DataFrame(dic)
print(df)

#to see specify col use columns---
var=pd.DataFrame(dic,columns=["name"])
print('\nSpecific col\n',var)   

#to get specific values from dataframe---
print('\nSpecific value--->',df["name"][5])

#to make dataframe using series---
list_df=pd.DataFrame({"s":pd.Series([1,2,3,4,5]),"w":pd.Series([11,22,33,44,55])})
print('\ndf from series--->\n',list_df) 
print('\n')
```

# Arithmetic operations

```python
var1=pd.DataFrame({"s":[12,34,12,23],"w":[45,23,55,23]})
print(var1)

#addition----

var1["v"]=var1["s"]+var1["w"]
print('\nadd--->\n',var1)

#substraction----

var1["v"]=var1["s"]-var1["w"]
print('\nsub--->\n',var1)

#Multiply----

var1["v"]=var1["s"]*var1["w"]
print('\nMultiple--->\n',var1)

#divide----

var1["v"]=var1["s"]//var1["w"]
print('\nDivide--->\n',var1)

#Logical data----returns true false 

var1["Py"]=var1["s"]<=var1["w"]
print('\nless than--->\n',var1)

var1["Py"]=var1["s"]>=var1["w"]
print('\nGreater than--->\n',var1)

var1["Py"]=var1["s"]!=var1["w"]
print('\nnot equal too--->\n',var1)
```

# Handling missing data--> dropna and fillna
* dropna is used to drop the NaN values.
* fillna is used to fill the NaN values.

## dropna

```python
csv_file=pd.read_csv("C:\\Nency\\SEM-4\\LAB\\PYTHON\\pandas\\Test_file2.csv")
print(csv_file)

#to drop a NaN value
print('\nto delete col--->\n',csv_file.dropna(axis=1))  
print('\nto delete row--->\n',csv_file.dropna(axis=0)) 

#to delete row with all values NaN--use 'how'

print('\nto delete row with all NaN--->\n',csv_file.dropna(how='all')) #it'll delete rows with all values NaN
print('\nto delete row with any value NaN--->\n',csv_file.dropna(how='any')) #it'll delete rows with any values of it NaN

#to delete rows with values NaN using col name--use 'subset'

print('\nto delete row with col name--->\n',csv_file.dropna(subset=["s"]))

#remove NaN using 'inplace'

print('\nto delete NaN values--->\n',csv_file.dropna(inplace=False)) 

#thresh--function specifies the minimum number of non-missing values (non-NaN) 
#eg: if thresh=2 then it'll keep rows with 2 or more values 


print('\nto delete NaN values using thresh--->\n',csv_file.dropna(thresh=3)) #here keeping rows which are having 3 values
```

## fillna

```python
csv_file=pd.read_csv("C:\\Nency\\SEM-4\\LAB\\PYTHON\\pandas\\Test_file2.csv")
print(csv_file)

#to fill a NaN value
print('\nto fill NaN--->\n',csv_file.fillna("heyy"))

#to fill NaN with different values acc to col name--
print('\nto fill NaN--->\n',csv_file.fillna({"s":"s","t":"t"}))

#to fill NaN with either forward or backward data--use 'method'
print('\nto fill NaN with forward--->\n',csv_file.fillna(method="ffill"))

#using inplace
#inplace wants either value or method with it

print('\nto fill NaN with inplace--->\n',csv_file.fillna(20,inplace=False))

#to fill NaN using limit---
#It's used to fill data upto set limit
#if limit=1 then it'll fill rows with 1 NaN value
```

## Join
* Here if col names are same than throws error require to change col name.

```python
dic1={"name":["ds","djs","dws","hdhdw","jwdww","ueuwue"],"roll no":[23,45,45,74,94,48]}
dic2={"xyz":["a","v","c","s"],"no":[223,452,145,743]}
df1=pd.DataFrame(dic1)
df2=pd.DataFrame(dic2)
print('join 1st and 2nd\n',df1.join(df2))
print('\njoin 2nd and 1st\n',df2.join(df1)) 

#to get whole data in joining 2nd with 1st--use 'how'
print('\nfull join 2nd and 1st\n',df2.join(df1,how="right")) #by default left==inner join and right==outer join

#if we use same col names than error.
#to resolve we can change col name during runtime using lsuffix and rsuffix.

ndf1=pd.DataFrame({"name":["ds","djs","dws","hdhdw","jwdww","ueuwue"],"roll no":[23,45,45,74,94,48]})
ndf2=pd.DataFrame({"xyz":["a","v","c","s"],"roll no":[223,452,145,743]})
print('\nfull join 2nd and 1st same changing col name\n',ndf2.join(ndf1,how="right",lsuffix='_1',rsuffix='_1'))
```

# Groupby
* Simply use to group the data according to the specified col name.

```python
df=pd.DataFrame({"name":["ds","ds","dws","dw","dw","ueuwue","a","a"],"roll no":[23,45,45,74,94,48,22,27]})
gdf=df.groupby("name")
print(gdf)

for x,y in gdf:
    print(x)
    print(y)
    print()

# to get specific col data--use 'get_group'
print('\nspecific data--->\n',gdf.get_group("ds"))

#to get minimum data from all the groups--use 'min'
print('\nminimum data--->\n',gdf.min())

#to get maximum data from all the groups--use 'max'
print('\nmaximum data--->\n',gdf.max())

#similarly for mean--use 'mean'

#to get data in list--use 'list'
```

# Melt
* It's used to change dataframe format from wide to long

```python
df=pd.DataFrame({"name":["ds","ds","dws","dw","dw","ueuwue","a","a"],"roll no":[23,45,45,74,94,48,22,27]})
print('without melt-->\n',df)
print('\nmelt-->\n',df.melt())

#to make any col id--use 'id_vars'
print('\nnew id-->\n',pd.melt(df,id_vars=["roll no"]))

#to change variable name--'var_name'
print('\nnew variable name-->\n',pd.melt(df,id_vars=["roll no"],var_name="xyz"))

#to change value name--'value_name'
print('\nnew value name-->\n',pd.melt(df,id_vars=["roll no"],var_name="xyz",value_name="pqr"))
```

# Pivot
*  Used to reshape a DataFrame by transforming unique values from one column into new columns
*  it requires to arg index and columns
*  basically format data acc to the specified col name

```python
df=pd.DataFrame({"name":["ds","ds","dws","dw","dw","dws","a","a"],"roll no":[23,45,45,74,94,48,22,27],"marks":[23,44,34,45,21,24,56,89]})
print(df.pivot(index="roll no",columns="name"))

#to get data to specific col only
print('\n',df.pivot(index="roll no",columns="name",values="roll no"))

#to make aggregate functions on table--use 'aggfunc'
print('\n agg func mean-->\n',df.pivot_table(index="name",columns="roll no",aggfunc="mean", margins=True))
#margins show overall result of the table

print('\n agg func sum-->\n',df.pivot_table(index="name",columns="roll no",aggfunc="sum", margins=True))

```
